```python
x=[100,134,245,356]
y=[166,277,388,499]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, marker='o')
plt.ylabel("P")
plt.xlabel("Q")
plt.show()
```

```python
plt.plot(x,y, '--')
plt.ylabel("P")
plt.xlabel("Q")
plt.show()
```

```python
plt.plot(x,y, '*')
plt.ylabel("P")
plt.xlabel("Q")
plt.show()
```

```python
plt.plot(x,y, '3')
plt.ylabel("P")
plt.xlabel("Q")
plt.show()
```

```python
plt.plot(x,y, 'r')
plt.ylabel("P")
plt.xlabel("Q")
plt.show()
```

```python
plt.plot(x,y, 'o:r')
plt.ylabel("P")
plt.xlabel("Q")
plt.show()
```

```python
plt.plot(x,y, 'o:r',ms =10)
plt.ylabel("P")
plt.xlabel("Q")
plt.show()
```

```python
plt.plot(x,y, 'o:r',ms =10, mec ='y',mfc='r')
plt.ylabel("P")
plt.xlabel("Q")
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[12,13,14,15,16,17]
z=[21,22,23,24,25,26]
```

```python
plt.plot(x,y, '*:r', ms=10,mec='y',mfc='r')
plt.plot(x,z, 'b:o',ms=10, mec='g', mfc='r')
plt.show()
```

```python
plt.plot(x,y, '*:r', ms=10,mec='y',mfc='r')
plt.plot(x,z, 'b:o',ms=10, mec='g', mfc='r')
plt.ylabel("y axis",fontdict={'color': 'blue','size':20})
plt.xlabel("x axis",fontdict={'color': 'green','size':21})
plt.show()
```

```python
plt.plot(x,y, '*:r', ms=10,mec='y',mfc='r')
plt.plot(x,z, 'b:o',ms=10, mec='g', mfc='r')
plt.ylabel("y axis",fontdict={'family':'serif','color': 'blue','size':20})
plt.xlabel("x axis",fontdict={'color': 'green','size':21})
plt.show()
```

```python
plt.plot(x,y, '*:r', ms=10,mec='y',mfc='r')
plt.plot(x,z, 'b:o',ms=10, mec='g', mfc='r')
plt.ylabel("y axis",fontdict={'family':'serif','color': 'blue','size':20})
plt.xlabel("x axis",fontdict={'color': 'green','size':21})
plt.title("graph")
plt.show()
```
